```python
import json
import openai
from IPython.display import Markdown
```

```python
# OpenAI API authentication
with open('secrets.json') as f:
    secrets = json.load(f)

if 'OPENAI_API_KEY' in secrets:
  print('OpenAI API keys not found.')
  
openai.api_key = secrets['OPENAI_API_KEY']
openai.Model.list()
```

```python
# Define model and prompt
model = 'gpt-3.5-turbo'
prompt = '''
Factors:
- Readability
- Uniformity
- Understandability
- Correctness
- Performance
- Security

Write a HTML/CSS code file that has issues related to some of the above factors and perform a code review on it to improve the code in those factors without updating the code itself. Your response must follow this template:

### Instruction:
{instruction to perform code review}

### Input:
{code file}

### Output:
- Readability (Status: {GOOD || POOR})
{list of code review}
- Uniformity (Status: {GOOD || POOR})
{list of code review}
- Understandability (Status: {GOOD || POOR})
{list of code review}
- Correctness (Status: {GOOD || POOR})
{list of code review}
- Performance (Status: {GOOD || POOR})
{list of code review}
- Security (Status: {GOOD || POOR})
{list of code review}
'''
```

```python
# Create chat completion
completion = openai.ChatCompletion.create(
  model=model,
  messages=[
    {'role': 'system', 'content': 'You are a helpful assistant.'},
    {'role': 'user', 'content': prompt},
  ]
)
```

```python
display(Markdown(completion.choices[0].message.content))
```
